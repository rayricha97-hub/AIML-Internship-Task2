```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Display summary statistics for numerical features
print("--- Summary Statistics for Numerical Features ---")
display(df.describe())

# Display summary statistics for categorical features
print("\n--- Summary Statistics for Categorical Features ---")
display(df.describe(include=['O']))
```

```python
plt.figure(figsize=(12, 5))

# 1. Histogram for Passenger Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True, color='skyblue', bins=20)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')

# 2. Histogram for Ticket Fare Distribution
plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], kde=True, color='salmon', bins=20)
plt.title('Fare Distribution of Passengers')
plt.xlabel('Fare')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# 1. Boxplot for Passenger Age
plt.subplot(1, 2, 1)
sns.boxplot(y=df['Age'], color='lightgreen')
plt.title('Boxplot of Passenger Age (Detecting Outliers)')

# 2. Boxplot for Ticket Fare
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Fare'], color='gold')
plt.title('Boxplot of Passenger Fare (Detecting Outliers)')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Calculate the correlation matrix for numerical features only
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

# Plot the correlation heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# 1. Analysis of Survival Count based on Gender
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Survived', hue='Sex', palette='Set2')
plt.title('Survival Count by Gender (0 = Died, 1 = Survived)')
plt.xticks([0, 1], ['Died', 'Survived'])

# 2. Analysis of Survival Count based on Passenger Class
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Survived', hue='Pclass', palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.xticks([0, 1], ['Died', 'Survived'])

plt.tight_layout()
plt.show()
```
